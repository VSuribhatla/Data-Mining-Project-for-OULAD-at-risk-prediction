```python
def weighted_sum(data, value, column):
    data[column] = data[value].cumsum()
    return data
```

```python
assessments_new = assessments.groupby(['code_module','code_presentation']).agg({'weight' : 'sum'})
assessments_new.reset_index(level=[0,1], inplace=True)
assessments_new.columns = ['code_module','code_presentation','total_marks']
```

```python
assessments_new = pd.merge(assessments, assessments_new, how='left', on=['code_module','code_presentation'])
assessments_new['total_percent'] = np.divide(assessments_new['weight'],assessments_new['total_marks'])*100
assessments_new.groupby(['code_module','code_presentation']).size()
assessments_new = assessments_new.groupby(['code_module','code_presentation']).apply(weighted_sum, 'total_percent', 'total_weight')
```

```python
assessments_new
```

```python
student_assessments_new = pd.merge(studentAssessment, assessments_new, how = 'outer', on='id_assessment' )
student_assessments_new
```

```python
student_assessments_new['score'].replace('?','0',inplace=True)
student_assessments_new['date'].replace('?','0',inplace=True)
student_assessments_new1['score'].replace('?','0',inplace=True)
student_assessments_new1['date'].replace('?','0',inplace=True)
```

```python
student_assessments_new.date_submitted = np.round(student_assessments_new.date_submitted).astype(int)
student_assessments_new.id_student = np.round(student_assessments_new.id_student).astype(int)
student_assessments_new['score'] = student_assessments_new['score'].astype(int)
student_assessments_new['date'] = student_assessments_new['date'].astype(int)
student_assessments_new1.date_submitted = np.round(student_assessments_new1.date_submitted).astype(int)
student_assessments_new1.id_student = np.round(student_assessments_new1.id_student).astype(int)
student_assessments_new1['score'] = student_assessments_new1['score'].astype(int)
student_assessments_new1['date'] = student_assessments_new1['date'].astype(int)
```

```python
student_assessments_new['submission'] = np.where(student_assessments_new['date'] < student_assessments_new['date_submitted'], 1, 0)
student_assessments_new1['outcome'] = np.where(student_assessments_new1['score'] >= 40, 1, 0)
student_assessments_new1['score_per_weights'] = np.divide(student_assessments_new1['score'],100)*student_assessments_new1['total_percent']
student_assessments_new1.dropna()
student_assessments_new1 = student_assessments_new1.groupby(['code_module','code_presentation','id_student']).apply(weighted_sum,'score_per_weights', 'score_for_total')
student_assessments_new1 = student_assessments_new1[['code_module','code_presentation','id_student','id_assessment','assessment_type','date','date_submitted','is_banked',
                                                   'score','weight','outcome','total_marks','total_percent','score_per_weights','total_weight','score_for_total']]
student_assessments_new1.dropna()
```
